Which albums stood out (AS IN THE BIGGEST)in terms of sales?

```python
## import libraries and packages
import pandas as pd

## mad math functions package
from scipy.stats import zscore 

## run the display code here
pd.options.display.float_format = '{:,.2f}'.format
```

```python
lps= pd.read_csv ("https://raw.githubusercontent.com/sandeepmj/datasets/main/1967_hits.csv")
lps.head (10)
```

```python
from scipy.stats import zscore 
```

```python
lps ["sales_zscore"]= zscore (lps ["sales"])
lps
```

```python
largest_lps_z_score = largest_dfs_z_score = lps.sort_values(by='sales_zscore', ascending=False).iloc[0]
largest_lps_z_score
```

```python
album_with_highestzscore= "SGT. PEPPER'S LONELY HEARTS CLUB BAND"
```

```python
dfs= pd.read_csv ("https://raw.githubusercontent.com/sandeepmj/datasets/main/most-streamed-2023.csv")
dfs.head (10)
```

```python
conversion_rate = 1500 
```

```python
dfs['album_sales'] = dfs['streams_2023'] //conversion_rate
```

```python
print (dfs)
```

```python
dfs["sales_zscore"]= zscore (dfs ["album_sales"])
dfs
```

```python
largest_dfs_z_score = dfs.sort_values(by='sales_zscore', ascending=False).iloc[0]
largest_dfs_z_score
```

```python
f"The album that stood out the most in their respective years is the {album_with_highestzscore} by the Beatles in 1967"
```
